```
from gensim import corpora, models, similarities
import jieba
import csv
from csv import writer
with open('/content/drive/My Drive/Precily/Text_Similarity_Dataset.csv','r') as f:
    reader = csv.reader(f, delimiter=",")
    for i, line in enumerate(reader):
      if i==0:
        with open('result.csv', 'a+', newline='') as write_obj:
          # Create a writer object from csv module
          csv_writer = writer(write_obj)
          # Add contents of list as last row in the csv file
          csv_writer.writerow(['Unique_ID','similarity_score'])

        continue
      else:

        text1 =line[1]
        text2 = line[2]
        text1 = [jieba.lcut(text) for text in text1]
        dictionary = corpora.Dictionary(text1)
        feature_cnt = len(dictionary.token2id)
        corpus = [dictionary.doc2bow(text) for text in text1]
        tfidf = models.TfidfModel(corpus) 
        kw_vector = dictionary.doc2bow(jieba.lcut(text2))
        index = similarities.SparseMatrixSimilarity(tfidf[corpus], num_features = feature_cnt)
        sim = index[tfidf[kw_vector]]
        l=[]
        for j in range(len(sim)):

          l.append(sim[j])
        # create a result.csv file 
        with open('result.csv', 'a+', newline='') as write_obj:
          # Create a writer object from csv module
          csv_writer = writer(write_obj)
          x=sum(l)/len(l)
          print('similarity of pair with Unique_ID ',i,' is:',x)
          # Add contents of list as last row in the csv file
          csv_writer.writerow([i,x])
```
